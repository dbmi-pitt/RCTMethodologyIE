```python
true_file_path  = "/home/lhoang2/Notebooks/RCT_Methodology_Extraction/DATA_07232022/RESULTS_07232022/true_labels_test_BERT.txt"

true_file = open(true_file_path, 'r')
true_lines = true_file.readlines()
true_test_labels = []
for true_line in true_lines:
    true_line = true_line.replace("\n","")
    true_line = true_line.replace("[","")
    true_line = true_line.replace("]","")
    true_line = true_line.replace(" ","")
    true_line = true_line.replace("'","")
    true_line = true_line.split(",")
    true_test_labels.append(true_line)
print(len(true_test_labels))

print (true_test_labels[:10])

unique_entities = []
#collect entity names
for item in true_test_labels:
    for label in item:
        label = label.replace("B-","")
        label = label.replace("I-","")
        if label not in unique_entities:
            unique_entities.append(label)
print (len(unique_entities))
unique_entities.remove('O')
print (unique_entities)
```
